## Forward Propagation

```python
# 3 edges
w = 3
a = 2 + 3*w
J = a**2
print(f"a = {a}, J = {J}")
```

## Back Propagation

```python
# 2 edges
a_epsilon = a + 0.001       # a epsilon
J_epsilon = a_epsilon**2    # J_epsilon
k = (J_epsilon - J)/0.001   # difference divided by epsilon
print(f"J = {J}, J_epsilon = {J_epsilon}, dJ_da ~= k = {k} ")
```

```python
# 2 edges using sympy (dJ_da, dJ_dw = dJ_da * da_dw)
bw, bJ, ba = symbols("w,J,a")
bJ = ba**2        # Define relationships
```

```python
dJ_da = diff(bJ,ba)
dJ_da
```

```python
dJ_da.subs([(ba,a)])
```

```python
ba = 2 + 3*bw
ba
```

```python
da_dw = diff(ba, bw)
da_dw
```

```python
dJ_dw = da_dw * dJ_da
dJ_dw
```
